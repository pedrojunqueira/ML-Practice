# Seaborn 

Testing Seaborn Visualisation

```python
import seaborn as sns
```

```python
sns.set_style('white')
```

```python
# Load dataset to train
tips = sns.load_dataset('tips')
    
```

```python
tips.head()
```

```python
tips.shape
```

```python
tips.info()
```

```python


sns.relplot(x='total_bill', y='tip', hue='day',
            
            data=tips)
```

```python
import matplotlib.pyplot as plt

fig,ax = plt.subplots(2, 1,  figsize=(6, 10))

sns.boxplot(data=tips,x='day',y='tip',ax=ax[0])
sns.boxplot(data=tips,x='day',y='total_bill',ax=ax[1])



```

```python
sns.pairplot(data=tips, hue="smoker");
```

```python
sns.distplot(tips['tip'],bins=50)
sns.distplot(tips['total_bill'],bins=50)

```

```python
sns.distplot(tips['total_bill'],bins=50)
```

```python

sns.relplot(x='total_bill', y='tip', hue='sex',
            
            data=tips)
```

```python
tips['%tip']=tips['tip']/tips['total_bill']
```

```python
sns.relplot(x='total_bill', y='%tip', hue='sex',
            
            data=tips)
```

```python
averagetip=tips['%tip'].mean()
```

```python
print('the average tip is {:.2f}'.format(averagetip))
```

```python
sns.lmplot(x='total_bill', y='%tip', data=tips,
           height=4)
```

```python
tips[tips['%tip']>.3]
```

```python
sns.lmplot(x='total_bill', y='%tip', data=tips.drop([67,172,178]),
           height=4)
```
